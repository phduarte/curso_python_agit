# Plotar Grafico de Pontos

```python
# entender os decimais como ',' e os milhares como '.'
cerveja = pd.read_csv('Consumo_cerveja.csv', decimal=b',', thousands='.', index_col=0, parse_dates=True)
```

```python
cerveja.head()
```

```python
# deletando valores faltantes
cerveja.dropna(inplace=True)
```

```python
cerveja.plot.scatter(x='Temperatura Maxima (C)', y='Consumo de cerveja (litros)', figsize=[12,8])
```

# Plotando Histograma

```python
cerveja['Temperatura Media (C)'].hist(figsize=[12,8])
```

# Plotando Grafico de Linha

```python
cerveja['Temperatura Media (C)'].plot(x='Data',figsize=[12,8])
```

```python
# Agrupando por mês
cerveja.groupby(pd.Grouper(freq='M')).mean()
```

```python
# transformando o index em coluna para usar como eixo x do grafico
cerveja.groupby(pd.Grouper(freq='M')).mean().reset_index()
```

```python
cerveja.groupby(pd.Grouper(freq='M')).mean().reset_index().plot(x='Data', y='Consumo de cerveja (litros)')
```

# Agrupando e Plotando

```python
# media de vendas de cerveja agrupado por dia de semana / final de semana
cerveja.groupby('Final de Semana')['Consumo de cerveja (litros)'].mean().plot.bar()
# substituir os rotulos do eixo x para Nao e Sim
plt.xticks(list(range(2)),['Nao','Sim'], rotation=0)
plt.show()
```

# Analisando se o numero de mortes é real ou provavelmente fake

```python
# Criando a régua do Benford
benford = np.zeros(9)
for i in range(9):
    benford[i]  = np.log10(i + 2) - np.log10(i + 1)
benford
```

```python
suicidios.suicides_no.astype(str).map(lambda x: x[0]).value_counts().drop('0')
```

```python
suicidios.suicides_no.astype(str).map(lambda x: x[0]).value_counts().drop('0').sum()
```

```python
real = (suicidios.suicides_no.astype(str).map(lambda x: x[0]).value_counts().drop('0') / suicidios.suicides_no.astype(str).map(lambda x: x[0]).value_counts().drop('0').sum()).values
```

```python
benford_suicidios = pd.DataFrame()
benford_suicidios['Esperado'] = benford
benford_suicidios['Observado'] = real
benford_suicidios.plot.bar(figsize=[12,8], title='Suicídios', )
plt.xticks(list(range(9)),list(range(1,10)))
plt.xlabel('Dígitos',labelpad=15)
plt.show()
```
